## Pure python implementation of a decision tree with requirements described above:

### To implement a decision tree, first we need the Attribute Selection Measures 


The following function is written separetely and is used by the fuctions which retrieves the gini index and entropy gain ratio. This function basically checks whether a feature name is valid

```
def isFeatureValid(dataset, feature):
    allFeatures = dataset.columns
    result = False
    for i in range(0, len(dataset.columns)):
        if allFeatures[i] == feature:
            result = True
            break
    return result
```

#### The following two functions are used to calculate the Gini Index and Gini Split

```
#Function to Calculate Gini index
def GiniIndex(dataset):
    outputSeries = dataset['output']
    uniqueOutputKeys = outputSeries.value_counts().keys().tolist()
    uniqueOutputValues = outputSeries.value_counts().tolist()
    result = 1    # initially the result is 1 because of the formula: GINI(T) = 1 - Σ [p(j|t)]^2
    
    for i in range(0, len(uniqueOutputKeys)):
        prob = uniqueOutputValues[i]/len(dataset['output'])
        result = result - (prob * prob)     #GINI(T) = 1 - Σ [p(j|t)]^2
    return result
    
#function to calculate the Gini Split
def GiniSplit(dataset, output, feature):
    if isFeatureValid(dataset, feature) == False:
        print("invalid column name!")
        return
    
    currentFeatureSeries = dataset[feature]    #list of all the features
    
    valueCounts = currentFeatureSeries.value_counts()
    valueCountsKeys = currentFeatureSeries.value_counts().keys().tolist()
    valueCountsValues = currentFeatureSeries.value_counts().tolist()
    
    fullDataToCheck = pd.DataFrame({'current_feature': currentFeatureSeries, 'output': output})
    result = 0
    for i in range(0, len(valueCountsKeys)):    #this loop sums all gini indexes by the formula GINIsplit = Σ (ni/n)*GINI(i)
        partitionedData = fullDataToCheck[fullDataToCheck['current_feature'] == valueCountsKeys[i]]
        gIndex = GiniIndex(partitionedData)
        
        result = result + (gIndex * len(partitionedData) / len(fullDataToCheck['output']))
    
    return result
```

The following two functions are used to calculate entropy and entropy gain ratio

```
# function to calculate entropy
def Entropy(dataset):    #this function calculates the entropy using the formula Entropy(T) = -Σ P(j|t) * log(P(j|t))
    outputSeries = dataset['output']
    uniqueOutputKeys = outputSeries.value_counts().keys().tolist()
    uniqueOutputValues = outputSeries.value_counts().tolist()
    result = 0
    
    for i in range(0, len(uniqueOutputKeys)):
        lgProb = uniqueOutputValues[i]/len(dataset['output'])
        result = result + (lgProb * math.log(lgProb, 2))    #P(j|t) * log(P(j|t))
    result = -1 * result    #at the end of the loop we need to negate the result because of the formula
    return result

#function to calculate entropy gain ratio
def EntropyGainRatio(dataset, output, feature):
    
    fullData = dataset
    fullData['output'] = output
    parentE = Entropy(fullData)    # parent entropy
    
    
    ## now find the entropy for all the childs that will be produced from using this feature
    if isFeatureValid(dataset, feature) == False:
        print("invalid column name!")
        return
    
    currentFeatureSeries = dataset[feature]
    valueCounts = currentFeatureSeries.value_counts()
    valueCountsKeys = currentFeatureSeries.value_counts().keys().tolist()
    valueCountsValues = currentFeatureSeries.value_counts().tolist()
    
    fullDataToCheck = pd.DataFrame({'current_feature': currentFeatureSeries, 'output': output})
    
    eGain = 0   #this variable stores the information gain, GAINsplit = Entropy(P) - Σ ((ni/n)*Entropy(i))
    splitI = 0  # this variable stores the splitInfo, splitInfo = -Σ ((ni/n)*log(ni/n))
    for i in range(0, len(valueCountsKeys)):    #this loop uses the function Entropy(dataset), and calculates eGain and SplitI
        partitionedData = fullDataToCheck[fullDataToCheck['current_feature'] == valueCountsKeys[i]]
        
        entropy = Entropy(partitionedData)
        entropyM = entropy * len(partitionedData) / len(fullDataToCheck)
        eGain = eGain + entropyM
        
        nByN = len(partitionedData) / len(fullDataToCheck)
        splitI = splitI + nByN * ((math.log(nByN, 2)))
        
    
    eGain = parentE - eGain # because of the formula GAINsplit = Entropy(P) - Σ ((ni/n)*Entropy(i))
    splitI = splitI * (-1.0) #GainRatio = GAINsplit/SplitInfo
    
    
    if splitI == 0.0:    #floating point precision handling
        splitI = 0.0000001
    gRatio = eGain / splitI
    
    return gRatio
```

#### The following function is the raw python implementation of decision tree

Since we can assume that the features we deal with will be binary, I have used a binary tree structure to store the output of the Decision Tree Generation function. The parameters treeTracker and treeTrackerNextCheck holds the tree information into 2 arrays that is generated from the function. The root of the binary tree starts from zero. Because of that, in the next couple of cells, the function is initially called with root value 0.

```
# the decision tree generator function
#treeTracker holds which feature to check at each iteration
#treeTrackerNextCheck holds the indices values and which direction to go for which feature value
def GenerateDecisionTree(Dataset, output, AttributeSelectionMeasures, treeTracker, treeTrackerNextCheck, root):
    
    if len(Dataset) == 0:
        return
    
    bestFeature = None
    valueForBestFeature = -1    # this variables hold the best feature to split with after checking all the features
    allFeatures = Dataset.columns  #list of all features
    gain = None  #this variable holds the gain value from the particular Attribute Selection Measure for the best feature
    
    for i in range(0, len(allFeatures)):    # this loop uses the Attribute Selection Measure to get the best feature and calculates the gain
        if AttributeSelectionMeasures == "Gini Index":    #if AttributeSelectionMeasure is gini index we call the gini index related function
            gain = GiniSplit(Dataset, output, allFeatures[i])
        elif AttributeSelectionMeasures == "Entropy Gain Ratio":   #if AttributeSelectionMeasure is gain ratio, we call the related function
            gain = EntropyGainRatio(Dataset, output, allFeatures[i])
        else:
            print("No such measure found named " + str(AttributeSelectionMeasures))  #otherwise AttributeSelectionMeasure is invalid
            return
        if gain > valueForBestFeature:
            valueForBestFeature = gain
            bestFeature = allFeatures[i]
    
    
    bestFeatureColumn = Dataset[bestFeature]
    uniqueValues = bestFeatureColumn.unique()
    
    combinedData = Dataset
    combinedData['output'] = output
    for i in range(0, len(uniqueValues)):    # this loop calculates the next node of the tree, here the function recursively calls itself to go deeper into the tree
        separatedData = combinedData[combinedData[bestFeature] == uniqueValues[i]]
        if len(separatedData) == 0:    #we reached an end
            treeTracker[root] = bestFeature
            treeTracker[root*2+1] = "end"
            treeTracker[root*2+2] = "end"
            continue
        if len(separatedData) == len(Dataset):    #we reached an end
            treeTracker[root] = bestFeature
            treeTracker[root*2+1] = "end"
            treeTracker[root*2+2] = "end"
            treeTrackerNextCheck[root*2+1] = separatedData.iloc[0]['output']#uniqueValues[i]
            treeTrackerNextCheck[root*2+2] = separatedData.iloc[0]['output']#uniqueValues[i]
            continue
        if i == 0:
            treeTracker[root] = bestFeature
            treeTrackerNextCheck[root*2+1] = uniqueValues[i]
            output = separatedData['output']
            separatedData = separatedData.drop('output', 1)
            GenerateDecisionTree(separatedData, output, AttributeSelectionMeasures, treeTracker, treeTrackerNextCheck, 2*root+1)
        else:
            treeTracker[root] = bestFeature
            treeTrackerNextCheck[root*2+2] = uniqueValues[i]
            output = separatedData['output']
            separatedData = separatedData.drop('output', 1)
            GenerateDecisionTree(separatedData, output, AttributeSelectionMeasures, treeTracker, treeTrackerNextCheck, 2*root+2)
        
        
    return

# this is the actual function that we call go get the decision tree, this one calls the function above to generate the tree
#you need to call the function using the dataframe containing the dependant variables(Dataset), the output, and the Attribute Selection Measure 
#treeTracker holds which feature to check at each iteration
#treeTrackerNextCheck holds the indices values and which direction to go for which feature value
def DecisionTree(Dataset, output, AttributeSelectionMeasures):
    treeTracker = [None] * 100000000
    treeTrackerNextCheck = [None] * 100000000
    GenerateDecisionTree(Dataset, output, AttributeSelectionMeasures, treeTracker, treeTrackerNextCheck, 0)  # 0 here is the root
    return treeTracker, treeTrackerNextCheck
```

#### The following functions are used to test the decision tree on test data

```
# checks the tree for only one row/data point
def GetDecisionTreeResult(treeTracker, treeTrackerNextCheck, oneDataPoint):
    root = 0
    while treeTracker[root] != "end":
        currentFeatureToTest = treeTracker[root]
        thatFeatureResult = oneDataPoint[currentFeatureToTest]
        
        if thatFeatureResult == treeTrackerNextCheck[root*2 + 1]:
            root = root*2 + 1
        elif thatFeatureResult == treeTrackerNextCheck[root*2 + 2]:
            root = root*2 + 2
        elif treeTracker[root*2+1] == "end" and treeTracker[root*2+2] == "end":
            break;
        
    return treeTrackerNextCheck[root*2+1]

#this function uses the function above in a loop to get the prediction for all rows in a dataframe
def TestDecisionTree(treeTracker, treeTrackerNextCheck, data):
    prediction = []#pd.Series()
    for i in range(0, len(data)):
        temp = data.iloc[i]
        tempResult = GetDecisionTreeResult(treeTracker, treeTrackerNextCheck, temp)
        prediction.append(tempResult)# = prediction.set_value(i, tempResult)
    
    return prediction#.tolist()
```
